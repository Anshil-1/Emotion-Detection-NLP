```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset load
df = pd.read_csv(
    r'D:\train.txt\train.txt',
    sep=';',
    header=None,
    names=['text', 'emotion']
)

```

```python
df['emotion'].unique()
```

```python
unique_emotions = df['emotion'].unique()

emotion_numbers = {}

i = 0
for emo in unique_emotions:
    emotion_numbers[emo] = i
    i += 1

df['emotion'] = df['emotion'].map(emotion_numbers)

print(emotion_numbers)
print(df.head())
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
import string

def remove_punc(txt):
    return txt.translate(str.maketrans('', '', string.punctuation))

df['text'] = df['text'].apply(remove_punc)
```

```python
print(df['text'].head())
```

```python
def remove_numbers(txt):
    new = ""

    for i in txt:
        if not i.isdigit():
            new = new + i

    return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
    new = ""

    for i in txt:
        if i.isascii():
            new += i

    return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
df.loc[5]['text']
```

```python
def remove(txt):
    words = txt.split()

    cleaned = []

    for i in words:
        if i not in stop_words:
            cleaned.append(i)

    return ' '.join(cleaned)
```

```python

df['text'] = df['text'].apply(remove)

df.loc[1]['text']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = CountVectorizer(ngram_range=(3,3))

X = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df['text'],
    df['emotion'],
    test_size=0.20,
    random_state=42
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

bow_vectorizer = CountVectorizer()

X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

nb_model = MultinomialNB()

nb_model.fit(X_train_bow, y_train)

pred_bow = nb_model.predict(X_test_bow)

print(accuracy_score(y_test, pred_bow))
```

```python
tfidf_vectorizer = TfidfVectorizer()

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

nb2_model = MultinomialNB()

nb2_model.fit(X_train_tfidf, y_train)
```

```python
y_pred = nb2_model.predict(X_test_tfidf)

print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Vectorization
bow_vectorizer = CountVectorizer()

X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

# Model
lr_model = LogisticRegression(max_iter=1000)

# Training
lr_model.fit(X_train_bow, y_train)

# Prediction
y_pred_bow = lr_model.predict(X_test_bow)

# Accuracy
print("BoW Accuracy:", accuracy_score(y_test, y_pred_bow))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Vectorization
tfidf_vectorizer = TfidfVectorizer()

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Model
lr2_model = LogisticRegression(max_iter=1000)

# Training
lr2_model.fit(X_train_tfidf, y_train)

# Prediction
y_pred_tfidf = lr2_model.predict(X_test_tfidf)

# Accuracy
print("TF-IDF Accuracy:", accuracy_score(y_test, y_pred_tfidf))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_tfidf))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_tfidf)

print(cm)
```
